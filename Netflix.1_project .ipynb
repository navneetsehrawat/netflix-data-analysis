```python
types = df['type'].value_counts()
plt.pie(types, labels=types.index, autopct='%1.1f%%', shadow=True)
plt.title('movies and tv shows', color='red', size=15)
plt.show()
```

```python
generes = df['listed_in'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=generes.values, y=generes.index, palette='viridis', orient='h')
plt.title('top 10 popular generes', color='r', size=20)
plt.xlabel('count', color='green', size=15)
plt.ylabel('generes', color = 'green', size=15)
plt.show()
```

```python
country = df['country'].value_counts().head(10)
sns.barplot(x=country.values, y=country.index, palette='viridis')
plt.title('top 10 country by releasing movies and series', size=20, color='blue')
plt.xlabel('count', size=15, color='red')
plt.ylabel('country name', size=15, color='red')
plt.show()
```

```python
rating = df['rating'].value_counts().head()
plt.pie(rating, labels=rating.index, autopct='%1.1f%%', shadow=True)
plt.title('top')
plt.show()
```

```python
title=df['title'].value_counts().head(10)
duration=df['duration'].value_counts().head(10)
sns.barplot(y=title.index, x=duration.values, palette='viridis', orient='h')
plt.title('top 10 longest movies and series', color='blue', size=20)
plt.xlabel('duration', color='red', size=15)
plt.ylabel('title', color='red', size=15)
plt.show()
```
